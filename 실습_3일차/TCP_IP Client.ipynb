# ======================
# TCP/IP 클라이언트 예제
# ======================
# 사용자가 입력한 메시지를 서버로 전송하고, 서버의 응답을 받아 출력하는 간단한 클라이언트입니다.
# =====================

```python
import socket
```

```python
#------------------------------
# 1. 서버 접속 정보 설정
#------------------------------
HOST = '192.168.0.46'
PORT = 58129

#------------------------------
# 2. 소켓 생성
#------------------------------
client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM) # AF_INET는 소켓이 통신할 수 있는 주소 유형 지정 -> IPv4 주소

#------------------------------
# 3. 서버에 연결 시도
#------------------------------
client_socket.connect((HOST, PORT))
print(f"서버 {HOST}:{PORT}에 연결 되었습니다.")
print("메세지를 입력하세요. 종료하려면 'exit'를 입력 \n")

#------------------------------
# 4. 메세지 송수신 루프
#------------------------------
while True:
    message = input("보낼 메세지: ")   # 사용자 입력 대기

    # exit 입력시 종료
    if message.lower() == 'exit':
        client_socket.sendall(message.encode())  # 서버에 종료 알림
        break
        
    client_socket.sendall(message.encode())

    # 서버로부터 응답 수신
    data = client_socket.recv(1024).decode()
    print(f"서버 응답 : {data} \n")

#------------------------------
# 5. 소켓 종료
#------------------------------
client_socket.close()
print("클라이언트 종료 완료")
```
