```python
import socket # 네트워크 통신을 기본 모듈
```

```python
#------------------------------
# 1. 서버 기본 설정
#------------------------------

HOST = '192.168.0.46'
PORT = 58129       # 0~65535 중 하나 선택, 다른 사람 또는 같은 포트 중복 X)

#------------------------------
# 2. 소켓 생성
#------------------------------
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

#------------------------------
# 3. IP와 포트를 소켓에 바인딩 (연결)
#------------------------------
server_socket.bind((HOST, PORT))# 서버가 클라이언트 요청을 받을 수 있도록 설정


#------------------------------
# 4. 클라이언트 연결 대기 시작
#------------------------------

server_socket.listen()

print(f"서버가 {HOST}:{PORT} 에서 연결 대기 중입니다...")

#------------------------------
# 5. 클라이언트가 정보가 맞다면 연결 수락 (accept)
#------------------------------
client_socket, addr = server_socket.accept()
print(f"클라이언트 {addr} 연결 완료")

#------------------------------
# 6. 클라이언트와 메세지 송수신 루프
#------------------------------

while True:
    data = client_socket.recv(1024).decode()    # bytes -> str
    if not data:
        print("데이터 수신 종료(클라이언트 연결 해제됨)")
        break
    # 종료 명령 감지
    if data.lower() == 'exit':
        print("클라이언트 종료 요청 수신")
        break

    #수신된 메세지 출력
    print(f"클라이언트 메세지: {data}")

    # 서버의 응답 생성
    reply = f"서버 응답: [{data}] 잘 받았습니다."

    # 클라이언트로 응답 전송 (문자열 -> bytes 변환)
    client_socket.sendall(reply.encode())
#------------------------------
# 7. 연결 종료
#------------------------------
client_socket.close()
server_socket.close()
print("서버 종료 완료")
```
